```python
df2 = df[["CpG_Coordinates", "methylated_r1", "total_value_r1","methylated_r2", "total_value_r2", "methylated", "total_value"]]
```

```python
df2
```

```python
df2 = df2.dropna()
```

```python
df2
```

```python
# Compute `cfdna_vrf` column
df2['cfdna_vrf'] = ((df2['methylated_r1'] + df2['methylated_r2']) / 2) / (df2['total_value_r1'] + df2['total_value_r2'])

# Compute `islet_vrf` column
df2['islet_vrf'] = df2['methylated'] / df2['total_value']
```
